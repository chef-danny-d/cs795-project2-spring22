### Helper functions

The following function is used in the regression model part of this report. It is used to calculate the R-squared value of the model.

```python
def get_best_score(grid):
    
    best_score = np.sqrt(-grid.best_score_)
    print(best_score)    
    print(grid.best_params_)
    print(grid.best_estimator_)
    
    return best_score
```

```python
def save_best_score(grid, name, df):
    
    print(grid.best_score_)
    print(grid.best_params_)
    print(grid.best_estimator_)
    print(grid.best_index_)
    res = pd.DataFrame(grid.cv_results_)
    res.to_csv(name + '.csv')

    pred = grid.predict(df)
    print(pred)

    return pred
```

```python
def print_cols_large_corr(df, nr_c, targ) :
    corr = df.corr()
    corr_abs = corr.abs()
    print (corr_abs.nlargest(nr_c, targ)[targ])
```

The following function is used to generate a heatmap correlation matrix for the given data-frame. The heatmap is generated using the combination of the matplotlib and the seaborn library. 

```python
def plot_corr_matrix(df, nr_c, targ) :
    
    corr = df.corr()
    corr_abs = corr.abs()
    cols = corr_abs.nlargest(nr_c, targ)[targ].index
    cm = np.corrcoef(df[cols].values.T)

    plt.figure(figsize=(nr_c/1.5, nr_c/1.5))
    sns.set(font_scale=1.25)
    sns.heatmap(cm, linewidths=1.5, annot=True, square=True, 
                fmt='.2f', annot_kws={'size': 10}, 
                yticklabels=cols.values, xticklabels=cols.values
    )
    plt.show()
```

#### Analyzing numeric features with a Pearson correlation coefficient > 0.5

```python
correlation_df = pd.DataFrame(columns=['Features', 'Correlation Coefficient', 'p-value'])
for x in range(len(li_plot_num_feats)):
  r, p = stats.pearsonr(df_train[li_plot_num_feats[x]],df_train[target])
  curerntFeat = li_plot_num_feats[x]

  correlation_df = correlation_df.append({
      'Features': curerntFeat, 
      'Correlation Coefficient': float("{0:.4f}".format(r)),
      'p-value': float("{0:.4f}".format(p))
    }, ignore_index=True)
  
correlation_df[correlation_df['Correlation Coefficient'] >= min_val_corr]
```
